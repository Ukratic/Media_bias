```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv('scraping/full_clean_lemma_topic.csv')
```

```python
data.tail(5)
```

```python
topic_journals = data.groupby(['topic','Journal'])['Date'].count().reset_index()
daily_topics = topic_journals.groupby(['topic','Journal'])['Date'].mean().reset_index().rename(columns = {'Date':'daily_topics'})
```

```python
px.line(daily_topics, x='topic', y='daily_topics', color='Journal', title='Average daily topics by Journal',markers=True,width=1200,height=600)
```

Most journals have a roughly similar distribution of topics. The most obvious exception is "Les Echos", with the highest amount articles in the "Economie" topic, which is not surprising considering it is first and foremost a financial newspaper.

```python
px.histogram(data, x='Journal', color='topic', barmode='group', title='Number of articles by Journal and topic',width=1200,height=500)
```

Articles on politics are prevalent for all journals, which is to be expected in this time of presidential elections.

```python
daily_topic_journals = data.groupby(['topic','Journal','Date'])['Article'].count().reset_index()
```

```python
for x in daily_topic_journals['Journal'].unique():
    fig = px.line(daily_topic_journals[daily_topic_journals['Journal']==x], x='Date', y='Article', color='topic', title=f'Number of articles by topic for {x}',
    width=1200,height=500,markers=True,facet_row='Journal')
    fig.show()
```

We can clearly see the work cycle of Les Echos, with a significant drop in unique articles on Sundays.

```python
number_topic_date = data.groupby(['topic','Date'])['Article'].count().reset_index()
```

February 23th has a higher count (cut out in the lineplot below) that should not be taken into account : It was the first day of scraping, thus naturally has more unique articles since there were no previous days with which to compare.


```python
px.line(number_topic_date[number_topic_date['Date']!='2022-02-23'], x='Date',y='Article', color='topic', title='Distribution of topics over time',
    width=1200,height=500,markers=True)
```

Most journals clearly have more articles around April 10 and 24 : 1st and 2nd round of the presidential elections.<br>
The topic of French Politics is almost always at the top, though International news (mostly about Ukraine) take the lead in February and March for some journals.
